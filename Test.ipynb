To remove the problem of over fitting for certain countries Im going to remove all countries with a count lower than 90


Also to remove the medical adancements affecting mortality rate i will only consider countries with 2000 year or over

AI - remove all data with countries that have a count lower than 90 and print dataframe

```python
# Get the list of countries with a count of 90 or more
countries_to_keep = country_counts[country_counts >= 90].index

# Filter the DataFrame to include only these countries
df = df[df['Country'].isin(countries_to_keep)]

# Display the first few rows of the filtered DataFrame
print("DataFrame after removing countries with count < 90:")
display(df.head())
```

AI - remove all data where Year is less than 2000 and print daata frame

```python
# Remove data where 'Year' is less than 2000
df = df[df['Year'] >= 2000]

# Display the first few rows of the filtered DataFrame
print("DataFrame after removing data where Year < 2000:")
display(df.head())
```

Now lets start with protien consumption - Im going to create a table with protient addded. but before that I will check if there are any countries that do not have a value in the protient section

Ai - look at the Diet Calories Animal Protein, Diet Calories Plant Protein and print the count of values =0 or the cell with no value given for each of these categoroies

```python
# Columns to check
protein_columns = ['Diet Calories Animal Protein', 'Diet Calories Plant Protein']

for col in protein_columns:
    # Count zero values
    zero_count = (df[col] == 0).sum()
    # Count missing values (NaN)
    nan_count = df[col].isnull().sum()

    print(f"Column: {col}")
    print(f"  Number of zero values: {zero_count}")
    print(f"  Number of missing values (NaN): {nan_count}")
    print("\n")
```

```python
# Remove rows where 'Diet Calories Animal Protein' or 'Diet Calories Plant Protein' have missing values
df_cleaned = df.dropna(subset=['Diet Calories Animal Protein', 'Diet Calories Plant Protein'])

# Display the first few rows of the cleaned DataFrame
print("DataFrame after removing rows with missing protein values:")
display(df_cleaned.head())
```

We are now going to add portien to see the nutrient vs mortality we look at protient because thats most needed for growth and thus affects life expexrancy

Ai - Add the values in diet calories plant protien and diet calories animal protient and make new columb titles total protien

```python
# Add 'Diet Calories Animal Protein' and 'Diet Calories Plant Protein' to create a new column 'total protein'
df_cleaned['total protein'] = df_cleaned['Diet Calories Animal Protein'] + df_cleaned['Diet Calories Plant Protein']

# Display the first few rows of the DataFrame with the new column
print("DataFrame with 'total protein' column:")
display(df_cleaned.head())
```

Now we looks at life expeactancy vs protien regardless of country

AI - make a scatter plot graph on total protien in x axis and life expecatncy on y axis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='total protein', y='Life Expectancy', data=df_cleaned, scatter_kws={'alpha':0.3})

# Add title and labels
plt.title('Life Expectancy vs. Total Protein Consumption')
plt.xlabel('Total Protein (Diet Calories Animal Protein + Diet Calories Plant Protein)')
plt.ylabel('Life Expectancy')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

We can clearly see there is a relation between the two categories. As in an increase in protien intake leades to higher life expectnacy.

But there is quite a bit of outliers im assuming its due to unhealthy amount of fats to test this thim gooing ti find the summary staistics of fat consumed and control for outliers and then create a new scatter plot and find OLS to see if its a preditable model

AI - give me summary sataitsics for diet in fat include mean, median, mode, inter quartile range highest value and lowest value

```python
# Get the 'Diet Calories Fat' column
fat_data = df_cleaned['Diet Calories Fat']

# Calculate summary statistics
mean_fat = fat_data.mean()
median_fat = fat_data.median()
mode_fat = fat_data.mode().tolist() # mode can return multiple values
min_fat = fat_data.min()
max_fat = fat_data.max()

# Calculate quartiles and IQR
Q1 = fat_data.quantile(0.25)
Q3 = fat_data.quantile(0.75)
iqr_fat = Q3 - Q1

# Print the results
print(f"Summary Statistics for 'Diet Calories Fat':\n")
print(f"  Mean: {mean_fat:.2f}")
print(f"  Median: {median_fat:.2f}")
print(f"  Mode: {mode_fat}")
print(f"  Lowest Value: {min_fat:.2f}")
print(f"  Highest Value: {max_fat:.2f}")
print(f"  Q1 (25th percentile): {Q1:.2f}")
print(f"  Q3 (75th percentile): {Q3:.2f}")
print(f"  Interquartile Range (IQR): {iqr_fat:.2f}")
```
