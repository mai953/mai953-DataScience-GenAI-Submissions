## Analyze the distribution of numerical features

### Subtask:
Visualize the distribution of numerical columns using histograms or box plots.


**Reasoning**:
Visualize the distribution of numerical columns using histograms.



```python
import matplotlib.pyplot as plt

numerical_cols = ['Age', 'SibSp', 'Parch', 'Fare']

for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    plt.hist(df[col].dropna(), bins=30) # Drop NaN values for plotting
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"Distribution of {col}")
    plt.show()
```

## Analyze the distribution of categorical features

### Subtask:
Visualize the distribution of categorical columns using bar plots.


**Reasoning**:
Identify categorical columns and create bar plots for each to visualize their distribution.



```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_cols = ['Survived', 'Pclass', 'Sex', 'Embarked']

for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=col, data=df)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title(f"Distribution of {col}")
    plt.show()
```
